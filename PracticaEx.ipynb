```python
import pandas as pd 
import numpy as np
```

```python
amazon = pd.read_csv("amazon.csv")
```

```python
amazon.head(10)
```

```python
amazon.columns
```

```python
cols = ['Name', 'Author', 'Genre']
```

```python
cols_validas = []
cols_many_levels = []
cols_one_level = []

size = amazon.shape[0]
umbral = 0.95
```

```python
for c in cols:
    nivel = amazon.loc[:,c].drop_duplicates().shape[0]
    if (nivel/size) > umbral: 
        cols_many_levels.append(c)
    elif nivel == 1:
        cols_one_level.append(c)
    else: 
        cols_validas.append(c)
```

```python
cols_validas
```

```python
cols_many_levels
```

```python
cols_one_level
```

```python
amazon.loc[:,cols_validas].drop_duplicates()
```

```python
amazon.loc[:,"Author"]
```

```python
levels = amazon.groupby(["Author"], as_index = False).agg({"User Rating": ["sum"]})
```

```python
levels.columns = levels.columns.droplevel(1)
```

```python
levels
```

```python
levels["Percent"] = levels["User Rating"]/size
```

```python
levels
```

```python
levels.describe()
```

```python
levels["Author-t"] = "Malo"
```

```python
levels.apply(lambda r: r["Author"] if r["Percent"] >0.008 else r["Author-t"], axis=1)
```

```python
levels.loc[:, ["Author", "Author-t"]]
```

```python
amazon = amazon.merge(levels, on=["Author"], how="inner")
```

```python
levels.to_csv("autores.csv", index=False)
```
