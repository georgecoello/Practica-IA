Paso 2: Matriz de Covarianza
===

```python
print(datos.cov())
```

```python
# C贸digo de visualizaci贸n
plt.figure(figsize = (7,7))
plt.scatter(datos["ingreso"], datos["horas_trabajadas"], marker = 8, s = 550, color = "purple", alpha = 0.5)
plt.xlabel("Ingreso", fontsize = 15)
plt.ylabel("Promedio de Horas Trabajadas", fontsize = 15)
plt.text(-2.2, 1.5, "Positiva", fontsize = 28, color = "red")
plt.show()
```

```python
from numpy.linalg import eig 
```

```python
# Valores Eigen (Eigenvalores y Eigenvectores)
valores, vectores = eig(datos.cov())
```

```python
vector_azul = vectores[:,0]
vector_rojo = vectores[:,1]

print("Eingenvector Rojo", vector_rojo, "Eigenvalor", valores[1])
print("Eingenvector Azul", vector_azul, "Eigenvalor", valores[0])
```

```python
# C贸digo de visualizaci贸n
plt.figure(figsize = (7,7))
plt.axes().set_aspect("equal")

# Graficando los datos
plt.scatter(datos["ingreso"], datos["horas_trabajadas"], marker = 8, s = 550, color = "purple", alpha = 0.5)

#Graficando los Eigenvectores
plt.quiver(0,0,
           vector_azul[0]/abs(vector_azul[0])*valores[0],
           vector_azul[1]/abs(vector_azul[1])*valores[0],
           color="red",angles="xy",scale_units="xy",scale=1,width=0.02
          )

plt.quiver(0,0,
           vector_rojo[0]/abs(vector_rojo[0])*valores[1],
           vector_rojo[1]/abs(vector_rojo[1])*valores[1],
           color="blue",angles="xy",scale_units="xy",scale=1,width=0.02
          )
        
plt.xlabel("Ingreso", fontsize = 15)
plt.ylabel("Promedio de Horas Trabajadas", fontsize = 15)
plt.show()
```

Paso 4: Proyecci贸n de los datos
===

```python
proyectados = pd.DataFrame(datos.values @ vectores.T,
              columns = ["ingreso","horas_trabajadas"])

# C贸digo de visualizaci贸n
plt.figure(figsize = (8,8))
plt.axes().set_aspect("equal")

# Datos rotados
plt.scatter(datos["ingreso"], datos["horas_trabajadas"], marker = 8, s = 550, color = "purple", alpha=0.5)

# Proyecci贸n de los datos en 1 dimensi贸n
plt.scatter(proyectados["ingreso"],[-2]*len(proyectados["horas_trabajadas"]), s=250, color = "red", alpha = 0.5)
plt.scatter([-3]*len(proyectados["horas_trabajadas"]), proyectados["horas_trabajadas"], s = 250, color = "blue", alpha = 0.5)
plt.show()
```

Paso 5: Selecci贸n de los componentes
===

```python
from sklearn.decomposition import PCA
```

```python
# Obtenci贸n de Componentes Principales 
pca = PCA()
datos = pca.fit_transform(datos)
```

```python
# C贸digo de visualizaci贸n
plt.figure(figsize=(7,7))
plt.axes().set_aspect("equal")
plt.scatter(datos[:,0],[2]*datos[:,0].size, s=250,color="red",alpha=0.5)
plt.scatter(datos[:,1],[1]*datos[:,1].size, s=250,color="blue",alpha=0.5)
plt.ylim((0.2 , 2.2))
plt.show()

print("Varianza explicada por cada componente: ", pca.explained_variance_)
print("Proporci贸n de varianza explicada por cada componente: ", pca.explained_variance_ratio_)
```
