```python
import pandas as pd 
import numpy as np
from datetime import date
```

```python
vegetales = pd.read_csv("cc.csv")
```

```python
vegetales
```

```python
vegetales["date"] = pd.to_datetime(vegetales["date"])
```

```python
vegetales.dtypes
```

```python
vegetales["year"] = vegetales["date"].dt.year
vegetales["month"] = vegetales["date"].dt.month
```

```python
vegetales
```

```python
vegetales = vegetales.sort_values(["category", "item", "variety", "date"], ascending = True)
```

```python
vegetales
```

```python
vegetales["row_number"] = vegetales.groupby(["category","item","variety","year","month"],as_index=False).cumcount()
```

```python
vegetales
```

```python
v2 = vegetales.loc[ vegetales["row_number"] == 0, ["category", "item", "variety", "date", "price"] ]
```

```python
v2
```

```python
v2["prev_price"] = v2.groupby(["category","item","variety"],as_index=False)["price"].shift(1)
```

```python
v2
```

shift 
===
es la funcion que permite retomar el valor anterior, y el parÃ¡metro 1 indica que se bajarÃ¡ el valor anterior
para saber el valor siguiente entonces se usa el parÃ¡metro -1
El valor del parÃ¡metro (1) va a depender de los saltos que querramos hacer, si se quieren saber 2 valores previos entonces se coloca shift(2)

Operaciones Lambda
===
Permite ir registro oir registro y poder ver las iteraciones que se tienen

Alt 126 ~

```python
v2 = v2.loc[~pd.isnull(v2["prev_price"]), :]
```

```python
v2
```

```python
v2.apply( lambda r: r["price"] - r["prev_price"], axis=1 )
```

```python
v2["dif_price"] = v2.apply( lambda r: r["price"] - r["prev_price"], axis=1 )
```
